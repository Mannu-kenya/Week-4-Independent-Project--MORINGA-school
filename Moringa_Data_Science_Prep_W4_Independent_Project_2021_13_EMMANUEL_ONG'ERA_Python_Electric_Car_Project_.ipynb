#Cleaning our data#

```python
#Here we start by dropping columns that we will not in our analysis


df.drop(["Displayed comment", "Scheduled at", "month", "year", "Cars"], axis = 1, inplace = True)

#We then preview our data and confirm if changes have been effected 

df
```

#Uniformity#

```python
#We rename columns and fix syntax errors

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('_counter', '').str.replace('1.4', '14')

df
```

#Removing Outliers#

```python
#dropping outliers using interquantile method

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df_without_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))]

df_without_outliers.head(5)
```

#Checking for NaN#

```python
#We check for Null values as follows

df_without_outliers.isnull().sum()
```

#Filling the null values#

```python
#We use the .bfill to backward fill the missing values in our dataset. 

new_filled_data= df_without_outliers.fillna(method='bfill') #This will fill the NaN values in our dataset

#Preview our datset to see if this has been effected

new_filled_data.head()
```

#QUESTIONS#

*1.0 Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018*

```python
new_filled_data['bluecar_diff'] = new_filled_data.bluecar.diff()
new_filled_data['utilib_diff'] = new_filled_data.utilib.diff()
new_filled_data['utilib_14_diff'] = new_filled_data.utilib_14.diff()


paris = new_filled_data['city'] == 'Paris'
pickup = new_filled_data['bluecar_diff'] < 0

paris_p_p_hours = new_filled_data[paris & pickup].groupby('hour')['hour'].value_counts().sort_values(ascending = False).head(1)
paris_p_p_hours
```

*2.0 What is the most popular hour for returning cars?*

```python
dropoff =new_filled_data['bluecar_diff'] > 0

new_filled_data[dropoff].groupby('hour')['hour'].value_counts().sort_values(ascending = False).head(1)
```

*3.0 What postal code is the most popular for picking up Blue cars?*

```python
pickup = new_filled_data['bluecar_diff'] < 0

new_filled_data[pickup].groupby('postal_code')['hour'].count().sort_values(ascending = False).head(1)
```
